```python
import time

class TimerDecorator:
    def __init__(self, arg=5):
        
        self.liczba_wywolan = 0
        self.czasy = []
        if callable(arg):
            self.func = arg
            self.max_history = 5 
        else:
            self.func = None
            self.max_history = arg
    def __call__(self, *args, **kwargs):
        
        if self.func is None:
            self.func = args[0]
            return self
        start = time.perf_counter() 
        
        wynik = self.func(*args, **kwargs)
        
        koniec = time.perf_counter()
        czas_trwania = koniec - start
        self.liczba_wywolan += 1
        self.czasy.append(czas_trwania)
        if len(self.czasy) > self.max_history:
            self.czasy.pop(0)
            
        return wynik

    def pobierz_statystyki(self):
        if self.liczba_wywolan == 0:
            return None
            
        return {
            "nazwa_funkcji": self.func.__name__, 
            "liczba_wywolan": self.liczba_wywolan,
            "najkrotszy_czas": min(self.czasy),
            "najdluzszy_czas": max(self.czasy),
            "sredni_czas": sum(self.czasy) / self.liczba_wywolan
        }

    def statystyki(self):
        dane = self.pobierz_statystyki()
        
        if dane is None:
            print(f"Funkcja '{self.__name__}' nie była jeszcze wywoływana.")
            return
            
        print(f"Statystyki dla funkcji: {dane['nazwa_funkcji']} \n")
        print(f"Liczba wywołań:  {dane['liczba_wywolan']}")
        print(f"Najkrótszy czas: {dane['najkrotszy_czas']:.6f} s")
        print(f"Najdłuższy czas: {dane['najdluzszy_czas']:.6f} s")
        print(f"Średni czas:     {dane['sredni_czas']:.6f} s")
        print("-" * 30)

@TimerDecorator(3)
def spanie(czas_snu):
    time.sleep(czas_snu)
    return f"Spano przez {czas_snu} s"

@TimerDecorator
def obliczenia(n):
    return sum(i**2 for i in range(n))


spanie(0.1)
spanie(0.2)
spanie(0.15)
spanie(0.1)
spanie(0.5)
print(spanie.czasy)
spanie.statystyki()

obliczenia(1000)
obliczenia(100000)
obliczenia(1000)
obliczenia(1000)
obliczenia(1000)
obliczenia(1000)
print(obliczenia.czasy)
obliczenia.statystyki()
```

```python
class Temperature:
    zero_absolutne_c = -273.15
    
    def __init__(self, celsius):
        self.celsius = celsius 

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        if value < self.zero_absolutne_c:
            self._celsius = self.zero_absolutne_c
        else:
            self._celsius = value

    @property
    def kelvin(self):
        return self._celsius - self.zero_absolutne_c

    @kelvin.setter
    def kelvin(self, value):
        if value < 0:
            value = 0
        self.celsius = value + self.zero_absolutne_c

    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

    @fahrenheit.setter
    def fahrenheit(self, value):
        celsius_val = (value - 32) * 5/9
        self.celsius = celsius_val

    @property
    def is_freezing(self):
        return self._celsius <= 0

    @property
    def is_boiling(self):
        return self._celsius >= 100

    @property
    def state(self):
        if self._celsius < 0:
            return 'solid'  
        elif self._celsius < 100:
            return 'liquid' 
        else:
            return 'gas' 

    def __repr__(self):
        return f"Temperature({self._celsius:.2f}°C)"

print("25 deg Celsius:")
t = Temperature(25)
print(f"Celsius: {t.celsius}")
print(f"Fahrenheit: {t.fahrenheit}")
print(f"Kelvin:     {t.kelvin}")
print(f"Stan wody:  {t.state}")

print("\n32 Fahrenheit:")
t.fahrenheit = 32
print(f"Celsius: {t.celsius}")
print(f"Freezing: {t.is_freezing}")
print(f"Stan wody: {t.state}")

print("\n373.15 Kelvin:")
t.kelvin = 373.15
print(f"Celsius: {t.celsius}")
print(f"Boiling: {t.is_boiling}")
print(f"State: {t.state}")

print("\n -300 Kelvin:")
t.kelvin = -300
print(f"Wynik Kelwinów zablokowany na: {t.kelvin}")
print(f"Wynik Celsjuszy:               {t.celsius}")

t
```

```python
import matplotlib.pyplot as plt

def lcg(seed, a, c, m, N):
    x = seed
    for _ in range(N):
        x = (a * x + c) % m
        yield x

def lcg_float(seed, a, c, m, N):
    
    generator_bazowy = lcg(seed, a, c, m, N)
    
    for x in generator_bazowy:
        yield x / m



print("petla for")
parametry = {'seed': 1, 'a': 5, 'c': 3, 'm': 16, 'N': 4}
generator = lcg(**parametry)

for i, liczba in enumerate(generator, start=1):
    print(f"{i}: {liczba}")


print("\n'next()'")
generator = lcg(seed=1, a=5, c=3, m=16, N=3)
print(f"1: {next(generator)}")
print(f"2: {next(generator)}")
print(f"3: {next(generator)}")


print("\nGenerator Float [0, 1)")
float_generator = lcg_float(seed=1, a=5, c=3, m=16, N=4)
for val in float_generator:
    print(f"{val:.5f}")

```
