```python
#zad 1

class counter:
    def __init__(self):
        self.value = 0  
        
    def up(self):
        self.value += 1
        
    def down(self):
        self.value -= 1
        
    def reset(self):
        self.value = 0
        
    def get_value(self):
        return self.value
    
c = counter()
c.up()
c.up()
print(c.get_value())  
c.down()
print(c.get_value())
c.reset()
print(c.get_value())
```

```python
#zad 2
from dataclasses import dataclass

@dataclass
class Measurement:
    time: str
    value: float
    unit: str

measurements = [
    Measurement("10:00", 22.5, "°C"),
    Measurement("10:05", 24.1, "°C"),
    Measurement("10:10", 21.8, "°C")
]

for m in measurements:
    print(f"{m.time}: {m.value}{m.unit}")


max_m = max(measurements, key=lambda x: x.value)
print(f"Największy pomiar: {max_m.value}{max_m.unit} o {max_m.time}")
```

```python
#zad 3
import math

class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self): return f"Vector2D[{self.x}, {self.y}]"
    def __add__(self, other): return Vector2D(self.x + other.x, self.y + other.y)
    def __mul__(self, scalar): return Vector2D(self.x * scalar, self.y * scalar)
    __rmul__ = __mul__
    
    def length(self): return math.sqrt(self.x**2 + self.y**2)
    def dot_product(self, other): return self.x * other.x + self.y * other.y


v1, v2 = Vector2D(2, 3), Vector2D(4, -1)
print(f"Suma: {v1 + v2}")
print(f"Długość v1: {v1.length():.2f}")
print(f"Iloczyn skalarny: {v1.dot_product(v2)}")
```

```python
#zad 4
class MeasurementSeries:
    def __init__(self, data): self.data = list(data)
    def __getitem__(self, index): return self.data[index]
    def __setitem__(self, index, val): self.data[index] = val
    def __len__(self): return len(self.data)
    def __iter__(self): return iter(self.data)
    
    def mean(self): return sum(self.data) / len(self.data) if self.data else 0
    def get_max(self): return max(self.data)
    def get_min(self): return min(self.data)


series = MeasurementSeries([10, 12, 15, 14])
series[0] = 11
print(f"Średnia: {series.mean()}")
print(f"Min: {series.get_min()}, Max: {series.get_max()}")
print("Pomiary:", list(series))
```

```python
#zad 5
class Polynomial:
    def __init__(self, coeffs):
        self.coeffs = coeffs
        
    def degree(self):
        max_deg = 0
        for p, c in enumerate(self.coeffs):
            if c != 0:
                max_deg = p
        return max_deg

    def __call__(self, x):
        return sum(c * (x ** p) for p, c in enumerate(self.coeffs))
        
    
    def __str__(self):
        czlon = []
        for p, c in enumerate(self.coeffs):
            if c != 0:
                if p == 0: czlon.append(f"{c}")
                elif p == 1: czlon.append(f"{c}x")
                else: czlon.append(f"{c}x^{p}")
        return " + ".join(czlon) if czlon else "0"

p = Polynomial([1, 0, 3])  

print(f"Równanie: P(x) = {p}")
print(f"Stopień: {p.degree()}")
print(f"P(2) = {p(2)}")                            
```

```python
#zad 6
class Account:
    def __init__(self, username, points=0):
        self.username = username
        self.points = points
        
    def add_points(self, amount):
        self.points += amount
        
    def subtract_points(self, amount):
        self.points -= amount
        
    def info(self):
        return f"{self.username}: {self.points} pkt"


class PremiumAccount(Account):
    def __init__(self, username, points=0, bonus=2):
        super().__init__(username, points)
        self.bonus = bonus
        
    def add_points(self, amount):
        points_with_bonus = amount * self.bonus
        super().add_points(points_with_bonus)
```

```python
#zad 7
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name, x=0, y=0):
        self.name = name
        self.x = x
        self.y = y
        
    @abstractmethod
    def sound(self):
        pass
        
    @abstractmethod
    def move(self):
        pass
        
    def info(self):
        return f"{self.name} is at ({self.x}, {self.y})"


class Dog(Animal):
    def sound(self):
        return "hał"
        
    def move(self):
        self.x += 2  


class Bird(Animal):
    def sound(self):
        return "ćwir"
        
    def move(self):
        self.x += 5  
        self.y += 5


zoo = [Dog("Adam"), Bird("Michał"), Dog("Ahmed")]

for animal in zoo:
    animal.move()
    print(f"{animal.name} says '{animal.sound()}' and is now at ({animal.x}, {animal.y})")


    
```
